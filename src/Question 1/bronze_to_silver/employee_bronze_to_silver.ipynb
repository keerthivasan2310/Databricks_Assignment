```python

from pyspark.sql.types import *
from pyspark.sql.functions import *
```

```python
# Custom schema
employee_schema = StructType([
    StructField("EmployeeID", IntegerType(), True),
    StructField("Name", StringType(), True),
    StructField("Age", IntegerType(), True),
    StructField("Salary", DoubleType(), True),
    StructField("DepartmentID", IntegerType(), True),
    StructField("CountryID", IntegerType(), True)
])

```

```python
# Read Bronze Data
employee_df = spark.read.format("csv") \
    .option("header", True) \
    .schema(employee_schema)\
    .load("/Volumes/workspace/default/assignment/source_bronze/employee/")
display(employee_df)
```

```python
employee_df = rename_columns_snake(employee_df)
```

```python
display(employee_df )
```

```python
employee_df = employee_df.withColumn("load_date", current_date())
```

```python
display(employee_df)
```

```python
spark.sql("CREATE DATABASE IF NOT EXISTS Employee_info")
```

```python
employee_df.write.format("delta") \
    .mode("overwrite") \
    .option("mergeSchema", "true")\
    .saveAsTable("dim_employee")
```
