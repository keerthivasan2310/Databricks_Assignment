```python
from pyspark.sql.functions import col
import re
```

```python
 # Function to convert CamelCase → snake_case
def camel_to_snake(col_name):
    s1 = re.sub('(.)([A-Z][a-z]+)', r'\1_\2', col_name)
    return re.sub('([a-z0-9])([A-Z])', r'\1_\2', s1).lower()

def rename_columns_snake(df):
    for c in df.columns:
        df = df.withColumnRenamed(c, camel_to_snake(c))
    return df
```
